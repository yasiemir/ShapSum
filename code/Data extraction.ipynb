```python
import glob
import os
```

```python
fname = glob.glob('../dataset/duc2002/data/test/summaries/duc2002abstracts/extracts_abstracts/d*')
```

```python
from HTMLParser import HTMLParser
# create a subclass and override the handler methods
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        pass
    def handle_endtag(self, tag):
        pass
    def handle_data(self, data):
        if len(data) > 5:
            text.append(data.strip())
# instantiate the parser and fed it some HTML
parser = MyHTMLParser()
```

```python
j = 1
for i in fname:
    f = os.path.join(i,'perdocs')
    print 'Extracting data from filename {}'.format(f[0:10]+str('....'))
    with open(f) as fp:
        docstring = fp.read()
        docstring = docstring.replace('\n',' ').replace('\r',' ')
        text = []
        parser.feed(docstring)
        string = ' '.join(text)
    with open('../multi/multi.a.{0:03d}.txt'.format(j),'w') as f:
        f.write(string)
        j += 1
    f_summary = os.path.join(i,'200')
    
```
